```python
import pandas as pd
import numpy as np
import math
import random
from sklearn.model_selection import train_test_split #pa regresion lineal
import collections
import operator
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
X = df.iloc[:, 0:4].values
y = df.loc[:,"Clase"].values
```

```python
#Utilizamos la función train_test_split de sklearn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

### Función que devuelve la distancia euclídea entre dos vectores

```python
def distancia_euclidea(a,b):
    suma =0
    for i in range(len(a)):
        suma+= (a[i]-b[i])**2
    return np.sqrt(suma)
```

```python
n=3 # Número de vecinos con los que se comparará cada patrón
clase=[]
for j in range(len(X_test)):
    patron_prueba = X_test[j]

    distancias_patron=[]
    for i in range(len(X_train)):
        distancias_patron.append( [distancia_euclidea(patron_prueba, X_train[i]), i] )

    #ordenar de menor a mayor las distancias al patrón
    distancias_patron.sort()
    #Obtener los índices de los patrones del X_train, que tienen las n distancias menores con el patrón
    indice_etiquetas = [distancias_patron[:n][i][1] for i in range(len(distancias_patron[:n]))]
    #Obtener las etiquetas de los índices anteiores
    etiquetas = [y_train[i] for i in indice_etiquetas]

    #dic diccionario cuyas claves son las etiquetas de los n patrones más cercanos a patron_prueba, claves: el número de veces que se repiten esas etiquetas
    dic = collections.Counter(etiquetas)
    #Ordenar dic por valores de mayor a menor
    #Obtener la clase que más se repite
    clase_patron = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)[0][0]
    #En clase se guarda la etiqueta asignada a cada patrón
    clase.append(clase_patron)
```

## df de prueba

```python
df_prueba = pd.DataFrame(X_test, columns=df.columns[0:4])
df_prueba["y_test"]=y_test
df_prueba["y_pred"]= clase
df_prueba["aciertos"] = np.where(df_prueba["y_test"] == df_prueba["y_pred"], 1, 0)
df_prueba["aciertos"].value_counts(normalize=True)
```
