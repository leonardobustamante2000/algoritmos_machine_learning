```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv("iris.csv")

#Función que devuelve la distancia euclídea entre dos vectores
def distancia_euclidea(a,b):
    suma =0
    for i in range(len(a)):
        suma+= (a[i]-b[i])**2
    return np.sqrt(suma)
```

#### Elegir centroides al azar sin repetir, como grupos tendrá el kmeans

```python
def primeros_centorides(n):
    indices=np.random.choice(range(0,150), n, False) 
    return [df.iloc[i, 0:4].values for i in indices]

```

### Calacular las distancias de cada punto a los centroides: 
    Estas distancias se guardan en dis_centroides junto con el indice del centroide. 
    Escogemos la distancia mínima, junto con el indice del centroide con el que resulta esa distancia mínima.
    Guardamos en etiquetas este índice


```python
def etiquetas_kmeans(centroides, df):

    dis_centroides=[]
    etiquetas=[]
    for j in range(len(df)):
        dis_centroides=[]
        for i in range(len(centroides)):
            dis_centroides.append([distancia_euclidea(centroides[i],  df.iloc[j, :-1].values),i])

        etiquetas.append(min(dis_centroides)[1])
    return etiquetas
```

### Segunda iteración:
    Se escogen como nuevos centroides las medias de las características en cada clase

```python

def nuevos_centroides(df):
    centroides=[]
    
    for i in df["clase"].value_counts().index.values:
        centroides.append(df[df["clase"]==i].iloc[:,0:4].describe().loc["mean", :].values)

    return centroides
```

## Función kmeans: Devuelve las etiquetas de los datos
### parámetro:
    df: el data frame que contiene  los datos
    n: número de clusters que se quieren formar
    i: Número de iteraciones del kmeans

```python
def k_means(df,n,i):

    #primera iteración
    centroides = primeros_centorides(n) # Se eligen al azar los primeros centroides sin repetirse, n número de clusters
    etiquetas = etiquetas_kmeans(centroides, df) # Calcula la distancia de cada punto a los centroides, devolviendo el indice del centroide que dió la menor distancia
    
    #Asignar las etiquetas a los valores, primer etiquetado    
    df["clase"]=etiquetas
    print(df["clase"].value_counts())

    for i in range(i-1):
        # Siguientes iteraciones 
        centroides = nuevos_centroides(df) # Devuelve como centroides las medias de las características de cada clase
        etiquetas = etiquetas_kmeans(centroides, df)
        
        #Asignar las etiquetas a los valores        
        df["clase"]=etiquetas
        print(df["clase"].value_counts())
        
    return etiquetas

```

### Con la función k_means se puede ver como, en cada iteración, va cambiando el número de elementos que pertenecen a cada clase, hasta que se estabiliza ese número

```python
k_means(df,3,10)
```
